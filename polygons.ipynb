```python
%matplotlib widget
import time
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation
from matplotlib.patches import Polygon
```

```python
def gen_vertices(n,r):
    vertices = []
    if n < 3:
        n = 3
    for i in range(n):
        angle = (2 * np.pi / n )* i
        x = r * np.cos(angle)
        y = r * np.sin(angle)
        vertices.append((x,y))
    return vertices

def animate():
    vertices = gen_vertices(3,1)
```

```python
# Vertices of the Koch snowflake
vertices = gen_vertices(8,1)

# Create a Polygon object
polygon = Polygon(vertices, closed=True, edgecolor='blue', facecolor='lightblue')

# Create a plot
fig, ax = plt.subplots()
ax.add_patch(polygon)

# Set axis limits
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.2, 1.2)

# Show the plot
plt.show()
```
